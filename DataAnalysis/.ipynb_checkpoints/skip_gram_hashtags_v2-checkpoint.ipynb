
# The Skip-gram model







## Word embeddings

In this nodebook we create **hashtag embeddings**, i.e., a mapping from hashtags to a vector space. These vectors can indicate semantic similarities between hashtags.
The two most popular algorithms for constructing word embeddings are **skip-gram** and **CBOW** (Continous Bag Of Words).
In both algorithms, word embeddings are learned through an auxiliary task. In CBOW, this task is predicting the current word from a window of **context** words, i.e., surrounding words. The auxiliary task under the skip-gram scheme is the opposite: given a current word, the task is predicting context words. In the domain of hashtags, we try to predict co-occuring hashtags given a hashtag. After training, we use learned weights from the model as our hashtag embeddings.

## Skip gram general architecture

*The global architecture of the base model of the skip-gram. Picture taken from [this website](https://www.geeksforgeeks.org/implement-your-own-word2vecskip-gram-model-in-python/)* 


skip-gram-2.jpg


We consider a vocabulary of hashtags $V$ for which we wish to create embeddings of dimension $N$, i.e., $\mathbf e_{wrd}(w) \in \mathbb{R}^N$. Additionally, we consider **all** hashtags in a post that co-occur in the set of hashtags for a post in the prediction step.

The first step in the algorithm for training a skip-gram model is the processing of each sample in our training data into a (multi-) set of pairs of input and context words ($w_i$, $w_t$), where $t\in \{1 \dots, i-1, i+1, \dots, M\}$ (where $M$ is the number of hashtags for a post).

We project the *one-hot* encoding of our input hashtag $\mathbf e_{oh}(w_i)\in \mathbb{R}^{|V|}$ with a weight matrix $W \in \mathbb{R}^{|V| \times N}$. This projection is then multiplied by weight matrix $W'$ and the softmax function is applied, resulting in a vector that represents the predicted probability of $w\in V$ appearing in the context of the $M-1$ other hashtags in a post. Our goal is to maximize the log-likelihood of our data: $\log p(w_t|w_i, W, W')$, which is typically done using gradient-descent methods.

At the end of training, we use the rows of matrix $W$ as our word embeddings, i.e., $\mathbf e_{wrd}(w_i) = W_i$.

Note: the columns of $W'$ will give us "context embeddings" $\mathbf e_{ctx}$, but we are only interested in the hashtag/word embeddings.



## Negative sampling

During training, our goal is to maximize $$p(w_t\mid w_i)=\frac{1}{Z(w_i)}\exp(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t)) = \frac{1}{Z(w_i)}\exp(W_i \cdot (W')^\top_t)  $$

However $Z(w)=\sum_{w' \in V} \exp(\mathbf e_{wrd}(w) \cdot \mathbf e_{ctx}(w'))$ can take a long time to compute for large $V$.
We can instead use **negative sampling**: rather than computing the $|V|$-dimensional probability distribution via the softmax, we use the sigmoid function between the ground truth label and the label of a word drawn uniformly at random from $V$. 

Formally, for each pair of words in our training data $(w_i,w_t)$, we randomly sample (with replacement) a set $C^-$ from $V$. Our objective under the negative sampling approach becomes

 $$\underset{W,W'}{\text{arg max}} \sum_{(w_i,w_t,C^-)} \left(  \log(p(w_t\mid w_i)) +  \sum_{w^- \in C^-}  \log(1-p(w^-|w_i)) \right)$$

where we compute $p(w_t\mid w_i)$ using the sigmoid function: $p(w_t|w_i) = \sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))$.


# Hyperparameters for creating the data set

```python
#size of the vocabulary
V_size = 12000

# appears max times
max_freq = 100000

# appears min times
min_freq = 2

# filter hashtags
filter_hashtags = []

#embedding size
N = 300

#number of negative samples per positive pairs (wt,wi)
K = 20
```

# Data loading and preprocessing

```python
import numpy as np
posts = np.genfromtxt('training_data_set.csv', delimiter=',', dtype=np.dtype(str), usecols=1)
# splits the string of hashtags
posts = np.char.split(posts)
print(posts)
```

```python
import string
def preprocessing(posts): 
    training_data = [] 
    for i in range(len(posts)): 
        post = posts[i]
        # remove punctuation
        x = [hashtag.strip(string.punctuation) for hashtag in post] 
        # make all hashtag lowercase
        x = [hashtag.lower() for hashtag in x] 
        if x:
          training_data.append(x) 
    return training_data 
posts = preprocessing(posts)
print(posts[:3])
```

```python
print("Number of posts:", len(posts))
```

# Prepare data for training


## Extract V

```python
# count how often hashtags appear all posts
from collections import defaultdict
count = defaultdict(int)
for post in posts: 
    for hashtag in post: 
      count[hashtag] += 1
```

```python
# sort hashtags by appearance frequency
import operator
sorted_counts = sorted(count.items(), key=operator.itemgetter(1), reverse=True)
```

Select the V most commons

```python
#print(sorted_counts)
```

```python
def check_valid_hashtag(pair):
    (hashtag, count) = pair
    if hashtag in filter_hashtags:
        return False
    elif count < min_freq:
        return False
    elif count > max_freq:
        return False
    else:
        return True

# filter hashtags
vocabulary = list(filter(check_valid_hashtag, sorted_counts))
```

## Convert the corpus into indices

Many of our operations can be vectorized if we represent our words with unique ids rather than the original string. 

We assign an id to all words and create lookup tables for word to index and index to word for fast lookup. For the rest of the problem, you will work with these indices.

```python
# Assign ids and create lookup tables
hashtag_to_index = {}
index_to_hashtag = []
for idx, tup in enumerate(vocabulary,0):
  hashtag_to_index[tup[0]] = idx
  index_to_hashtag.append(tup[0])
```

```python
assert len(index_to_hashtag) == len(hashtag_to_index)
print("Number of hashtag (unfiltered):", len(sorted_counts))
print("Number of hashtag (unfiltered):", len(index_to_hashtag))
```

Now we can transform our dataset by replacing the words with their index.

```python
posts_index = []
for post in posts:
  ids=[]
  for hashtag in post:
    # only add hashtags that are in the vocabulary!!! (all others dropped)
    if hashtag in hashtag_to_index:
      ids.append(hashtag_to_index[hashtag])
  posts_index.append(ids)
print("First three posts in represented by the indexes of their hashtags:")
print(posts_index[:3])
```

```python
print("Number of posts (after indexing them):", len(posts_index))
print("Number of hashtags (including duplications)", sum([len(x) for x in posts_index]))
```

## Extract pair (hashtag, context)

Create a list of training samples $C$: for each hashtag $w_i$ in the corpus, add a tupple ($w_i$, $w_t$) for each context hashtag $w_t$. Remember context hashtags are hashtags $w_t$ that co-occur with $w_i$.

```python
from tqdm.notebook import trange, tqdm
import matplotlib.pyplot as plt

# initalizing the training samples (an array containing one array per hashtag in the vocabulary with all context-word-pairs)
training_samples = [[]] * len(index_to_hashtag)

# used for discriptive statistics (to check that it works)
count = 0
counts = []

# iterate through all posts
for post in tqdm(posts_index):
    interim_count = 0
    #iterate through all hashtags of a post
    for i in range(len(post)):
        hashtag = post[i]
        # iterate through the context of that hashtag
        for j in range(0, len(post)):
            if j != i:
                interim_count += 1
                context_hashtag = post[j]
                
                # add context-hashtag-pair to the training samples
                if len(training_samples[hashtag]) == 0:
                    training_samples[hashtag] = [(hashtag, context_hashtag)]
                else:
                    training_samples[hashtag].append((hashtag, context_hashtag))
    count += interim_count
    counts.append(interim_count)
# displays the number of context-hashtag-pairs per post as histogram
fig = plt.figure()
plt.hist(counts)
fig.suptitle("Histogram of the number of training samples/context-hashtag-pairs per post:")
plt.xlabel('Number of training samples')
plt.ylabel('Number of posts')
plt.show()
```

```python
# Total number of context-word-pairs (training samples)
print("Number of training samples/context-hashtag-pairs:", sum([len(x) for x in training_samples]))
print("Manual count of training samples to validate:", count)
```

## Negative sampling

For each positive sample ($w_i$, $w_t$) in $C$, we create a set $C^-$ of $K$ negative samples (as defined in the beginning). We sample a hashtag proportional to its number of occurences (frequency) in the corpus and the negative samples are different from the positive context word $w_t$. We store all the training data as a list of tuple ($w_i$, $w_t$, $C^-$).

```python
import random

# initalize array to capture training samples
training_samples_incl_neg = [[]] * len(index_to_hashtag)

# filter out all hashtags from the posts that are not in the vocabulary to get the frequency of all hashtags appearing in the corpus
all_hashtag_rep = list(filter(lambda x: x in index_to_hashtag, [inner for outer in posts for inner in outer]))

# iterate through the array of arrays with the context-hashtag-pairs (training samples)
for hashtag_samples_ind in tqdm(range(len(training_samples))):
    hashtag_pairs_and_neg = []
    # iterate through the array with the context-hashtag-pairs (done for each word in the vocabulary)
    for sample in training_samples[hashtag_samples_ind]:
        neg_samples = []
        # repeat for K negative samples
        for i in range(K):
            same_as_context = True
            # while the randomly chosen sample (by chosing a random hashtag in the filtered set of all posts) is equal to the context hashtag, 
            # we choose a new one, else we add it to the list of negative samples.
            while same_as_context:
                neg = all_hashtag_rep[random.randint(0, len(all_hashtag_rep)-1)]
                neg_ind = hashtag_to_index[neg]
                same_as_context = neg_ind == sample[1]
            neg_samples.append(hashtag_to_index[neg])
        # create a tuple (w_i, w_t, C) where C = [(w_0^-, ..., w_20^-)] for every context-hashtag-pair
        hashtag_pairs_and_neg.append(sample + (neg_samples,))
    training_samples_incl_neg[hashtag_samples_ind] = hashtag_pairs_and_neg
```

# Learning

## Calculate gradient

Recall our training objective under the negative sampling approach. Under this objective, we can define our loss function with respect to our training data $\mathcal{D}$ as:

$$L(W,W') = - \sum_{(w_i,w_t,C^-) \in \mathcal{D}}\left (\log(\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))) +  \sum_{w^- \in C^-}  \log(1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w^-)))\right )$$

We will consider **Stochastic Gradient Descent** for training. This means at each iteration we consider only one datapoint $(w_i,w_t,C^-)$, we compute the loss and the gradients with this datapoint and update the parameters $e_{wrd}(w_i),e_{ctx}(w_t)$ and all the $e_{ctx}(w^-)$ for $w^- \in C^-$

We write down the partial derivative of the loss with respect to the hashtag embedding $\mathbf e_{wrd}(w_i)$ and context embedding $\mathbf e_{ctx}(w_t)$. Additionally, we provide the partial derivative of the loss with respect to a single negative sample $w^- \in C^-$: $\nabla_{\mathbf e_{wrd}(w_i)} L$, $\nabla_{\mathbf e_{ctx}(w_t)} L$, $\nabla_{\mathbf e_{ctx}(w^-)} L$


For clarity and to simplify the notation, we calculate the gradients with respect to a tuple $(w_i, w_t, C^-)$ (each step in the SDG later on). The loss is:

$$L(W, W') = - \sum_{(w_i,w_t,C^-) \in \mathcal{D}} L(W, W';w_i, w_t, C^-)$$
$$L(W, W';w_i, w_t, C^-) = - \left(\log(\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))) +  \sum_{w^- \in C^-}  \log(1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w^-)))\right)$$

Note that:
$$\nabla_{\mathbf e_{wrd}(w_i)} L = \sum_{(w_i,w_t,C^-) \in \mathcal{D}} \frac{\partial L(W, W';w_i, w_t, C^-)}{\partial e_{wrd}(w_i)}$$
$$\nabla_{\mathbf e_{ctx}(w_t)} L = \sum_{(w_i,w_t,C^-) \in \mathcal{D}} \frac{\partial L(W, W';w_i, w_t, C^-)}{\partial e_{ctx}(w_t)}$$ 
$$\nabla_{\mathbf e_{ctx}(w^-)} L = \sum_{(w_i,w_t,C^-) \in \mathcal{D}} \frac{\partial L(W, W';w_i, w_t, C^-)}{\partial e_{ctx}(w^{-})}$$

Partial derivative of the loss with respect to the hashtag embedding $\mathbf e_{wrd}(w_i)$:

$\frac{\partial L(W, W';w_i, w_t, C^-)}{\partial e_{wrd}(w_i)}$

$= \frac{\partial }{\partial e_{wrd}(w_i)} \left(- \left (\log(\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))) +  \sum_{w^- \in C^-}  \log(1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w^-)))\right )\right)$

$= - \left( \frac{\partial}{\partial e_{wrd}(w_i)} \log(\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))) + \frac{\partial }{\partial e_{wrd}(w_i)} \sum_{w^- \in C^-}  \log(1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w^-)))\right)$

$= - \left( \frac{\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))(1-\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t)))\cdot \mathbf e_{ctx}(w_t)}{\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))} + \sum_{w^- \in C^-}  \frac{-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w^-))(1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w^-)))\cdot e_{ctx}(w^-)}{1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w^-))} \right)$

$= (\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))-1)\cdot \mathbf e_{ctx}(w_t) + \sum_{w^- \in C^-}  \sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w^-))\cdot e_{ctx}(w^-)$

Partial derivative of the loss with respect to the hashtag embedding $\mathbf e_{ctx}(w_t)$:

$\frac{\partial L(W, W';w_i, w_t, C^-)}{\partial e_{ctx}(w_t)}$

$= \frac{\partial }{\partial e_{ctx}(w_t)} \left(\left (\log(\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))) +  \sum_{w^- \in C^-}  \log(1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w^-)))\right )\right )$

$= - \left( \frac{\partial}{\partial e_{ctx}(w_t)} \log(\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t)))\right )$

$= - \left( \frac{\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))(1-\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t)))\cdot \mathbf e_{wrd}(w_i)}{\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))}\right )$

$= (\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))-1)\cdot \mathbf e_{wrd}(w_i)$

Partial derivative of the loss with respect to the hashtag embedding $\mathbf e_{ctx}(w^-)$:

$\frac{\partial L(W, W';w_i, w_t, C^-)}{\partial e_{ctx}(w^{-})}$

$= \frac{\partial }{\partial e_{ctx}(w^{-})} \left(- \left (\log(\sigma(\mathbf e_{wrd}(w_i) \cdot \mathbf e_{ctx}(w_t))) +  \sum_{w_j^- \in C^-}  \log(1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w_j^-)))\right )\right )$

$= - \left(\frac{\partial }{\partial e_{wrd}(w_i)} \sum_{w_j^- \in C^-}  \log(1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w_j^-)))\right)$

$= - \left(\sum_{\underset{w_j^-=w^-}{w_j^- \in C^-}}  \frac{-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w_j^-))(1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{wrd}(w^i)))\cdot e_{ctx}(w_j^-)}{1-\sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w_j^-))} \right)$

$= \sum_{\underset{w_j^-=w^-}{w_j^- \in C^-}}  \sigma(\mathbf e_{wrd}(w_i) \cdot e_{ctx}(w_j^-))\cdot e_{wrd}(w^i)$

Consequently, the formula of the gradient update of $\mathbf e_{wrd}(w_i)$, $\mathbf e_{ctx}(w_t)$ and $\mathbf e_{ctx}(w^-)$ are:

$e_{wrd}(w_i)^{new} = e_{wrd}(w_i)^{old} + \eta \cdot \frac{\partial L}{\partial e_{wrd}(w_i)^{old}} = e_{wrd}(w_i)^{old} - \eta \cdot \sum_{(w_i,w_t,C^-) \in \mathcal{D}}\left( (\sigma(\mathbf e_{wrd}(w_i)^{old} \cdot \mathbf e_{ctx}(w_t)^{old})-1)\cdot \mathbf e_{ctx}(w_t)^{old} + \sum_{w^- \in C^-}  \sigma(\mathbf e_{wrd}(w_i)^{old} \cdot e_{ctx}(w^-)^{old})\cdot e_{ctx}(w^-)^{old} \right)$

$e_{ctx}(w_t)^{new} = e_{ctx}(w_t)^{old} + \eta \cdot \frac{\partial L}{\partial e_{ctx}(w_t)^{old}} = e_{ctx}(w_t)^{old} - \eta \cdot \sum_{(w_i,w_t,C^-)^{old} \in \mathcal{D}}\left( (\sigma(\mathbf e_{wrd}(w_i)^{old} \cdot \mathbf e_{ctx}(w_t)^{old})-1)\cdot \mathbf e_{wrd}(w_i)^{old}\right )$

$e_{ctx}(w^-)^{new} = e_{ctx}(w^-)^{old} + \eta \cdot \frac{\partial L}{\partial e_{ctx}(w^-)^{old}} = e_{ctx}(w^-)^{old} - \eta \cdot \sum_{(w_i,w_t,C^-) \in \mathcal{D}}\left(\sum_{\underset{w_j^-=w^-}{w_j^- \in C^-}}  \sigma(\mathbf e_{wrd}(w_i)^{old} \cdot e_{ctx}(w_j^-)^{old})\cdot e_{wrd}(w^i)^{old} \right)$

## Hyperparameters for training

```python
#training params
lr = 0.1
epochs = 5
```

## Training

Now we implement the gradient descent algorithm with the samples we created previously. The matrices W and W' are initialized with random weights between -0.8 and 0.8.

```python
# Helper function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
from sklearn.preprocessing import normalize
```

```python
import numpy as np
np.random.seed(42)
random.seed(42)

# vectorization of the training samples
vectorized_training_samples = [inner for outer in training_samples_incl_neg for inner in outer]

# initialization of weights to be between -0.8 and 0.8
W = np.random.rand(len(vocabulary), N).astype(np.float128)
W_prime = np.random.rand(N, len(vocabulary)).astype(np.float128)
W = (2*W-1)*0.8
W_prime = (2*W_prime-1)*0.8

W = normalize(W, axis=1, norm='l2')

epoch_losses = []
step_losses = []

# iterate through the number of epochs
for i in range(epochs):
    print("Epoch", i+1)
    epoch_loss = 0
    count = 0
    random.shuffle(vectorized_training_samples)
    t = tqdm(vectorized_training_samples, desc = "loss: {:.4f}".format(epoch_loss))
    
    # iterate through all samples of the training set
    for sample in t:
        wi = sample[0]
        wt = sample[1]
        C_minus = sample[2]
        
        # get the embedding of the hashtag and the context hashtags
        e_wi = W[wi]
        e_wt = W_prime[:,wt]
        
        # temporary variable to sum up the prdouct between the embedding of the hashtag 
        # and the sigmoid of the dot product of the embedding of the context hashtags and the hashtag
        s = 0
        # temporary variable to sum up the step loss
        step_loss = 0
        
        # iterate through negative samples
        for wm in C_minus:
            # get embedding of the negative sample
            e_wm = W_prime[:, wm]
            # update the weight of the context matrix for the (negative) sampled hashtag using GD
            W_prime[:, wm] = e_wm - lr/(i+1) * sigmoid(np.dot(e_wi, e_wm)) * e_wi
            # add to the temporary variable as described above
            s += sigmoid(np.dot(e_wi, e_wm)) * e_wm
            # add to the step loss
            t_step_loss = 1-sigmoid(np.dot(e_wi, e_wm))
            # case distinction for numberical stability
            if t_step_loss <= 0:
                step_loss -= np.log(10**-10)
            else:
                step_loss -= np.log(t_step_loss)
        
        # update weights of the hashtag embeddings
        W[wi] = e_wi - lr/(i+1) * ((sigmoid(np.dot(e_wi, e_wt)) - 1) * e_wt + s)
        
        # update weights of the context hashtag
        W_prime[:,wt] = e_wt - lr/(i+1) * (sigmoid(np.dot(e_wi, e_wt)) - 1) * e_wi
        
        # add to step loss
        step_loss -= np.log(sigmoid(np.dot(e_wi, e_wt)))
        epoch_loss += step_loss
        step_losses.append(step_loss)
        
        # for bookkeeping and updating loss
        count += 1
        if epoch_loss/count == np.inf:
            print(count)
        if count % 1000 == 0:
            t.set_description("loss: {:.8f}".format(epoch_loss/count))
            t.refresh()
    W = normalize(W, axis=1, norm='l2')
    epoch_loss = epoch_loss/len(vectorized_training_samples)
    print("Loss", epoch_loss)
    epoch_losses.append(epoch_loss)
    
# Plot loss during the epochs
fig = plt.figure()
plt.plot(range(epochs), epoch_losses)
fig.suptitle("Loss progression")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
# save embedding weights
np.savetxt('embedding.csv', W, delimiter=',')
```

```python
# plot losses during the individual context-word-negative-sample-tuples
fig = plt.figure()
plt.plot(range(len(step_losses)), step_losses)
fig.suptitle("Loss progression")
plt.xlabel('Training steps')
plt.ylabel('Loss')
plt.show()
```

```python
# plot loss averaged over 1000 successive context-word-negative-sample-tuples
step_losses_thousands = []
sum = 0
for i in range(len(step_losses)):
    sum += step_losses[i]
    if i % 1000 == 0:
        step_losses_thousands.append(sum/1000)
        sum = 0

fig = plt.figure()        
plt.plot(range(len(step_losses_thousands)), step_losses_thousands)
fig.suptitle("Loss progression")
plt.xlabel('Training steps in 1000s')
plt.ylabel('Loss')
plt.show()
```

## Cosine similarity calculation

```python
def cos_sim(e_x, e_y):
    vec_dot = np.dot(e_x, e_y)
    norm = np.linalg.norm(e_x)*np.linalg.norm(e_y)
    return vec_dot/norm
```

One way to compare vectors is the cosine similarity, which is defined as :
$$sim(\mathbf x,\mathbf y) = \frac{\mathbf x \cdot \mathbf y}{||\mathbf x|| ||\mathbf y||}$$

# Evaluation

```python
# calculates the average of the post's hashtags' embeddings
# expects list of hashtags
def post_avg_embedding(post):
    in_vocab = 0
    avg_vec = np.zeros(N)
    for hashtag in post:
        if hashtag in index_to_hashtag:
            ind_hashtag = hashtag_to_index[hashtag]
            avg_vec += W[ind_hashtag]
            in_vocab += 1
    if in_vocab == 0:
        raise Error('Post has 0 hashtags that have pretrained embeddings.')
    return avg_vec/in_vocab
```

```python
# calculates similarity between posts
# expects two lists of hashtags
def post_sim(post1, post2):
    vec1 = post_avg_embedding(post1)
    vec2 = post_avg_embedding(post2)
    return cos_sim(vec1, vec2)
```

```python
# calculates the average similarity between all post in the feed
# expects list of lists of hashtags
def feed_sim(feed):
    pairs = 0
    avg_sim = 0
    for i in range(len(feed)):
        post1 = feed[i]
        for j in range(i+1, len(feed)):
            post2 = feed[j]
            avg_sim += post_sim(post1, post2)
            pairs += 1
    return avg_sim/pairs
```
